```python
#KNN Classifier Model

x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
ClassifierModel_KNN = Pipeline([('knn_classifier',KNeighborsClassifier(n_neighbors=5,weights='distance'))])
```

```python
ClassifierModel_KNN.fit(x_train, y_train)
```

```python
#Train Score

ClassifierModel_KNN.score(x_train,y_train)
```

```python
#Test Score

ClassifierModel_KNN.score(x_test,y_test)
```

Justification:

As we can KNN results **accuracy** of **0.46** which is less than 50. As dataset is large KNN is not performing well and it may not work well if the data is noisy or if there are irrelevant features.

```python
#Train Score

ClassifierModel_svm.score(x_train,y_train)
```

```python
#Test Score

ClassifierModel_svm.score(x_test,y_test)
```
